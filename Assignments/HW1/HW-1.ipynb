```python
## Number 6 ################################

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Excel file
file_path = '../Assignments/HW1/IPA.xlsx'
IPA = pd.read_excel(file_path)

# Display the first few rows and basic info
print(IPA.head())
print(IPA.info())


# Calculate Summary statistics
summary_stats = IPA.groupby('Weeks')['IBU'].agg(['mean', 'std', 'median', 'min', 'max', 'count']).reset_index()
print("Summary Statistics:")
print(summary_stats)

# Visualizations
plt.figure(figsize=(10, 6))

# Boxplot
sns.boxplot(x='Weeks', y='IBU', data=IPA, order=['3 Weeks', '4 Weeks', '5 Weeks'], palette='Set2', hue='Weeks', legend=False)
plt.title('IBU Levels by Aging Time')
plt.xlabel('Aging Time')
plt.ylabel('International Bitterness Units (IBU)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.savefig('ibu_boxplot.png')

# Means plot with error bars
plt.figure(figsize=(10, 6))
sns.pointplot(x='Weeks', y='IBU', data=IPA, order=['3 Weeks', '4 Weeks', '5 Weeks'], capsize=.1)
plt.title('Mean IBU Levels by Aging Time (with 95% CI)')
plt.xlabel('Aging Time')
plt.ylabel('Mean IBU')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.savefig('ibu_means_plot.png')
```

```python
## Number 7 #################################################

import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
from scipy import stats
import matplotlib.pyplot as plt

# Excel file
file_path = '../Assignments/HW1/IPA.xlsx'
IPA = pd.read_excel(file_path)

# ANOVA model
model = ols('IBU ~ C(Weeks)', data=IPA).fit() # Changed 'df' to 'IPA'
anova_table = sm.stats.anova_lm(model, typ=2)

# Residuals for normality testing
residuals = model.resid

# Shapiro-Wilk
shapiro_stat, shapiro_p = stats.shapiro(residuals)

#  Q-Q Plot
plt.figure(figsize=(10, 6))
sm.qqplot(residuals, line='45', fit=True)
plt.title('Normal Q-Q Plot of Residuals')
plt.savefig('normality_qqplot.png')

print("ANOVA Table:")
print(anova_table)
print("\nShapiro-Wilk Test Result:")
print(f"Statistic: {shapiro_stat}, p-value: {shapiro_p}")
```

```python
## Question 8 ########################

import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

# Excel file
file_path = '../Assignments/HW1/IPA.xlsx'
IPA = pd.read_excel(file_path)

# Build the ANOVA model
model = ols('IBU ~ C(Weeks)', data=IPA).fit()

# 1. Visual Method: Residuals vs Fitted Values Plot
plt.figure(figsize=(10, 6))
plt.scatter(model.fittedvalues, model.resid)
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.title('Residuals vs Fitted Values')
plt.savefig('homogeneity_residual_plot.png')

# 2. Testing Method: Levene's Test
# Grouping data for the test
groups = [IPA[IPA['Weeks'] == w]['IBU'] for w in IPA['Weeks'].unique()]
levene_stat, levene_p = stats.levene(*groups)

print(f"Levene's Test Result:")
print(f"Statistic: {levene_stat}, p-value: {levene_p}")
```
